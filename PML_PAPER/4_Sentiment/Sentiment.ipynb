```python
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import pandas as pd

# 데이터 불러오기
file_path = '/content/drive/MyDrive/PML_PAPER/data/1_2_Processed_SuperSol.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# content 컬럼에서 텍스트 데이터를 추출
df = df.dropna(subset=['content'])  # content가 없는 행 제거
texts = df['content'].values  # 텍스트 데이터

# score 컬럼을 기준으로 라벨링 (1, 2, 3은 부정, 4, 5는 긍정)
df['label'] = df['score'].apply(lambda x: 0 if x in [1, 2, 3] else 1)  # 1,2,3은 1(positive), 4,5는 0(negative)

# TF-IDF 벡터화
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(texts)

# SVM 모델 정의 (RBF 커널 사용, 클래스 불균형 해결을 위해 class_weight='balanced' 설정)
svm_model = SVC(kernel='rbf', class_weight='balanced', probability=True)

# 모델 학습
svm_model.fit(X, df['label'])

# 예측: 감성 점수 (positive class의 확률을 추출)
df['sentiment_score'] = svm_model.predict_proba(X)[:, 1]  # 긍정 클래스(1)의 확률을 sentiment_score에 저장

# 예시로 전체 데이터에 대해 SVM 모델 평가
svm_y_pred = svm_model.predict(X)

# 평가 지표 출력
svm_accuracy = accuracy_score(df['label'], svm_y_pred)
svm_precision = precision_score(df['label'], svm_y_pred, average='weighted')  # 다중 클래스일 경우 weighted 방식
svm_recall = recall_score(df['label'], svm_y_pred, average='weighted')
svm_f1 = f1_score(df['label'], svm_y_pred, average='weighted')

print(f"Accuracy: {svm_accuracy}")
print(f"Precision: {svm_precision}")
print(f"Recall: {svm_recall}")
print(f"F1 Score: {svm_f1}")
print(classification_report(df['label'], svm_y_pred))

# 결과 확인: sentiment_score 컬럼 추가된 데이터프레임 출력
print(df.head())

# 새로운 CSV 파일로 저장
output_file_path = '/content/drive/MyDrive/PML_PAPER/data/4_Sentiment_SuperSol.csv'
df.to_csv(output_file_path, index=False)

# 저장 경로 출력
print(f"감성 점수가 포함된 파일이 저장되었습니다: {output_file_path}")
```
