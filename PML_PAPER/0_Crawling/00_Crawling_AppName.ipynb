## Review crawling

```python
from google_play_scraper import Sort, reviews_all

result = reviews_all(
    '[TargetApp]',
    sleep_milliseconds=0, # defaults to 0
    lang='ko', # defaults to 'en'
    country='kr', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # dafaults to Sort.MOST_RELEVANT
    filter_score_with=None
)
```

```python
import pandas as pd
df = pd.DataFrame(result)
```

```python
rev_df = df[['reviewId', 'content', 'score', 'thumbsUpCount', 'at', 'appVersion']]
```

```python
rev_df.info()
```

The content and appVersion columns contain null values.

```python
rev_df['content'].isnull().sum()
```

```python
# Remove rows with empty content
del_rev_df = rev_df.dropna(subset=['content'])
len(del_rev_df)
```
